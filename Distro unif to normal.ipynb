```python
def generador_muestra(a, b, n):
    if a < b:
        muestra = np.random.uniform(a, b, n)
    else:
        muestra = np.random.uniform(a, b, n)
    #print (muestra)
    #historigrama(s)
    return muestra
```

```python
def historigrama(s):
    count, bins, ignored = plt.hist(s, 'auto', density=True)
    plt.plot(bins, np.ones_like(bins), linewidth=2, color='r')
    plt.title("Muestra de distribuciÃ³n uniforme")
    plt.ylabel('frequencia')
    plt.xlabel('valores')
    plt.show()
```

```python
def prueba_normalidad(s, alpha=0.05):
    D, pval = ss.kstest(s, 'norm')
    print(D, pval)
    
    if pval < alpha:
        print('rechazado')
        return False
    
    else:
        print('aceptado')
        return True
    
```

```python
muestra = generador_muestra( 1, 10, 100 )
historigrama(muestra)
prueba_normalidad(muestra, 0.05)
```
